# Filtering Data with condition

````python
df = pd.read_csv(r"sales_data.csv")
````

````python
df.shape
````

````python
# check columns
df.columns
````

````python
filt = df['Customer_Age'] > 25
````

````python
df[filt]
````

````python
# with loc you can specific column you want of that particular criteria data
df.loc[filt, ['Country', 'Customer_Gender']]
````

# ```value_counts()```
> handy method to overview specific field of data but note that it only works with ```Series()``` not X```DataFrame```

````python
# set criterea to variable
profitFilt = df['Profit'] < 100
````

````python
# use loc to be able to filter out a single column or Series type so value_counts could be applied
df.loc[profitFilt, 'Country'].value_counts()
````
